# Converting DateTime column to 'datetime'

```python
df["DateTime"] = pd.to_datetime(df["DateTime"], format='mixed')
```

# Extracting 'Year', 'Month', 'Day', 'Hour' etc.. from each datetime column and creating a new column. These new columns are required for EDA (Exploratory Data Analysis).

```python
df['Year'] = df['DateTime'].dt.year
```

```python
df['Month'] = df['DateTime'].dt.month
```

```python
df['Day'] = df['DateTime'].dt.day
```

```python
df['Hour'] = df['DateTime'].dt.hour
```

```python
df['Weekday'] = df['DateTime'].dt.day_name()
```

```python
df['WeekOfYear'] = df['DateTime'].dt.isocalendar().week
```

```python
df['Minute'] = df['DateTime'].dt.minute
```

# Converting Month column to Season using U.S climate conditions

```python
def month_to_seasons(month):
    if month in [3, 4, 5]:
        season = 'Spring'
    elif month in [6, 7, 8]:
        season = 'Summer'
    elif month in [9, 10, 11]:
        season = 'Fall'
    elif month in [12, 1, 2]:
        season = 'Winter'
    return season
```

```python
df["Season"] = df["Month"].apply(month_to_seasons)
```

# Converting Hours column to Time-of-the-day 

```python
def hours_to_timing(hour):
    if hour in [21,22,23,0]:
        timing = 'Night'
    elif hour in range(0,12):
        timing = 'Morning'
    elif hour in range(12,18):
        timing = 'Afternoon'
    elif hour in range(18,21):
        timing = 'Evening'
    else:
        timing = 'X'
    return timing
```

```python
df["Time_of_Day"] = df["Hour"].apply(hours_to_timing)
```

# Calculating the percentage of readings for each month, preapred the dataframe for seaborn plotting

```python
month_counts = df['DateTime'].dt.to_period('M').value_counts(normalize=True).sort_index() * 100
month_rd = np.round(month_counts, decimals=1)

month_df = month_rd.reset_index()
month_df.columns = ['Month', 'Percentage']

month_df['Month'] = month_df['Month'].astype(str)

plt.figure(figsize=(5, 3))

sns.barplot(x='Month', y='Percentage', data=month_df, color='green', label='Percentage')

sns.lineplot(x='Month', y='Percentage', data=month_df, color='red', marker='o', label='Trend')

plt.title('Percentage of Readings by Month')
plt.xlabel('Month')
plt.ylabel('Percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Performing Multivariant Analysis by Filtering the DataFrame for 'In' and 'Out' places separately

```python
in_month = df[df['Place'] == 'In']['DateTime'].dt.to_period('M').value_counts(normalize=True).sort_index() * 100
in_month = np.round(in_month, decimals=1)
out_month = df[df['Place'] == 'Out']['DateTime'].dt.to_period('M').value_counts(normalize=True).sort_index() * 100
out_month = np.round(out_month, decimals=1)

in_out_month = pd.DataFrame({
    'Month': in_month.index,
    'In': in_month.values,
    'Out': out_month.values
})

in_out_month_melted = pd.melt(in_out_month, id_vars=['Month'], value_vars=['In', 'Out'], var_name='Place', value_name='Percentage')

hv.Bars(in_out_month_melted, ['Month', 'Place'], 'Percentage').opts(
    title="Monthly Percentage Readings by Place Variables",
    width=700,
    height=400,
    tools=['hover'],
    show_grid=True,
    ylabel="Percentage",
    legend_position='right'
)
```

```python
in_season = df[df['Place'] == 'In']['Season'].value_counts(normalize=True).sort_index() * 100
in_season = np.round(in_season, decimals=1)
out_season = df[df['Place'] == 'Out']['Season'].value_counts(normalize=True).sort_index() * 100
out_season = np.round(out_season, decimals=1)


in_out_season = pd.DataFrame({
    'Season': in_season.index.astype(str),
    'In': in_season.values,
    'Out': out_season.values
})


in_out_season_melted = pd.melt(in_out_season, id_vars=['Season'], value_vars=['In', 'Out'], var_name='Place', value_name='Percentage')


bars = hv.Bars(in_out_season_melted, kdims=['Season', 'Place'], vdims=['Percentage']).opts(
    title="Seasonal Percentage Readings by Place",
    width=700,
    height=400,
    tools=['hover'],
    show_grid=True,
    ylabel="Percentage",
    legend_position='right'
)

bars
```

```python
in_timing = df[df['Place'] == 'In']['Time_of_Day'].value_counts(normalize=True).sort_index() * 100
in_timing = np.round(in_timing, decimals=1)
out_timing = df[df['Place'] == 'Out']['Time_of_Day'].value_counts(normalize=True).sort_index() * 100
out_timing = np.round(out_timing, decimals=1)

in_out_timing = pd.DataFrame({
    'Time_of_Day': in_timing.index.astype(str),
    'In': in_timing.values,
    'Out': out_timing.values
})

in_out_timing_melted = pd.melt(in_out_timing, id_vars=['Time_of_Day'], value_vars=['In', 'Out'], var_name='Place', value_name='Percentage')

bars = hv.Bars(in_out_timing_melted, kdims=['Time_of_Day', 'Place'], vdims=['Percentage']).opts(
    title="Time_of_Day Percentage Readings by Place",
    width=700,
    height=400,
    tools=['hover'],
    show_grid=True,
    ylabel="Percentage",
    legend_position='right'
)

bars
```
